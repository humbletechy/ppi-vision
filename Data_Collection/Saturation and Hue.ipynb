```python
import PIL.Image
import matplotlib.pyplot as plt
import torch
from torchvision import transforms
```

```python
img = PIL.Image.open('/home/mehul/Downloads/washing_machine.jpeg')
```

```python
#Saturation
loader_transform3 = transforms.ColorJitter(saturation=2)
img3 = loader_transform3(img)
plt.imshow(img3)
plt.show()
```

```python
#Hue
loader_transform4 = transforms.ColorJitter(hue=0.2)
img4 = loader_transform4(img)
plt.imshow(img4)
plt.show()
```
